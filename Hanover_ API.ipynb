# Question: Optimal price to put in for bid on a house
Greater Richmond Area Cities ML model to best predict bid price based on house characteristics

Heatmap of selling prices based on selling price parsed by zip code Hover text with name, avg selling price, etc

Parameter Limits: 1 bathroom, 1 bedroom, PropertyType = Single Family, Townhouse, Condo, Duplex-Triplex, Manufactured

```python
# Import the dependencies.
import pandas as pd
import requests
import json

# Import API key
from config import API_KEY
```

```python
# API setup

url = "https://realty-mole-property-api.p.rapidapi.com/saleListings"

querystring = {"zipCode":"23111","limit":"500"}

headers = {
	"X-RapidAPI-Key": API_KEY,
	"X-RapidAPI-Host": "realty-mole-property-api.p.rapidapi.com"
}

response_HanoverA = requests.request("GET", url, headers=headers, params=querystring).json()

print(json.dumps(response_HanoverA, indent=4, sort_keys=True))
```
