```python
# Import dependencies
import pandas as pd
from pathlib import Path
import tensorflow as tf
```

```python
# Import our input dataset
attrition_df = pd.read_csv('train.csv')
print(attrition_df.shape)
attrition_df.head()
```

```python
# Count rows with NaN values
attrition_df.isnull().sum()
```

```python
# Remove rows with null values
attrition_df = attrition_df.dropna()
print(attrition_df.shape)
attrition_df.head(10)
```

```python
# Count unique variables in "Company Type" column

pd.unique(attrition_df["Company Type"])
```

```python
# Remove employee ID (not needed) + make date of joining a datetype datatype
attrition_df = attrition_df.drop(columns=["Employee ID"])
attrition_df.head()
```

```python
# Change date datatype to datetime
attrition_df["Date of Joining"]=pd.to_datetime(attrition_df['Date of Joining'])
```

```python
# write a for loop that loops through variables with object datatypes and encode them
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encoded_df = attrition_df.copy()

for col in attrition_df:
    if encoded_df[col].dtypes == "object":
        encoded_df[col] = le.fit_transform(encoded_df[col])

encoded_df.head()
```

```python
encoded_df.dtypes
```

```python
# Find range of values in continuous data columns

for col in encoded_df:
    if encoded_df[col].dtypes == "float64":
        print(f"The range for {col} is {encoded_df.min(axis=0)[col]}-{encoded_df.max(axis=0)[col]}")
```
